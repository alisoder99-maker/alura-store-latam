Análisis 1: ingreso total por tienda

```python
ingresos = {}

for nombre, df in tiendas.items():
    ingresos[nombre] = df['Precio'].sum()

ingresos
```

```python
df_ingresos = pd.DataFrame(list(ingresos.items()), columns=['Tienda', 'Ingreso Total'])
df_ingresos
```

```python
df_ingresos = df_ingresos.sort_values(by='Ingreso Total', ascending=False)
df_ingresos
```

Análisis 2: ventas por categoría

```python
ventas_categoria = {}

for nombre, df in tiendas.items():
    conteo = df['Categoría del Producto'].value_counts()
    ventas_categoria[nombre] = conteo
    print(f'--- {nombre} ---')
    print(conteo)
    print()
```

```python
for nombre, df in tiendas.items():
    categoria_top = df['Categoría del Producto'].value_counts().idxmax()
    cantidad_top = df['Categoría del Producto'].value_counts().max()
    print(f'{nombre}: categoría más vendida = {categoria_top} ({cantidad_top} ventas)')
```

```python
for nombre, df in tiendas.items():
    categoria_min = df['Categoría del Producto'].value_counts().idxmin()
    cantidad_min = df['Categoría del Producto'].value_counts().min()
    print(f'{nombre}: categoría menos vendida = {categoria_min} ({cantidad_min} ventas)')
```

Análisis 3: calificación promedio

```python
calificaciones = {}

for nombre, df in tiendas.items():
    calificaciones[nombre] = df['Calificación'].mean()

df_calificaciones = pd.DataFrame(list(calificaciones.items()), columns=['Tienda', 'Calificación Promedio'])
df_calificaciones = df_calificaciones.sort_values(by='Calificación Promedio', ascending=False)
df_calificaciones
```

```python
df_calificaciones['Calificación Promedio'] = df_calificaciones['Calificación Promedio'].round(2)
df_calificaciones
```

Análisis 4: productos más y menos vendidos

```python
for nombre, df in tiendas.items():
    conteo_productos = df['Producto'].value_counts()

    producto_mas_vendido = conteo_productos.idxmax()
    cantidad_mas_vendido = conteo_productos.max()

    producto_menos_vendido = conteo_productos.idxmin()
    cantidad_menos_vendido = conteo_productos.min()

    print(f'--- {nombre} ---')
    print(f'Más vendido: {producto_mas_vendido} ({cantidad_mas_vendido} ventas)')
    print(f'Menos vendido: {producto_menos_vendido} ({cantidad_menos_vendido} ventas)')
    print()
```

```python
resumen_productos = []

for nombre, df in tiendas.items():
    conteo = df['Producto'].value_counts()
    resumen_productos.append({
        'Tienda': nombre,
        'Producto más vendido': conteo.idxmax(),
        'Cantidad más vendido': conteo.max(),
        'Producto menos vendido': conteo.idxmin(),
        'Cantidad menos vendido': conteo.min()
    })

df_productos = pd.DataFrame(resumen_productos)
df_productos
```

```python
for nombre, df in tiendas.items():
    conteo = df['Producto'].value_counts()
    max_ventas = conteo.max()
    min_ventas = conteo.min()

    mas_vendidos = conteo[conteo == max_ventas]
    menos_vendidos = conteo[conteo == min_ventas]

    print(f'--- {nombre} ---')
    print('Más vendidos:')
    print(mas_vendidos)
    print('Menos vendidos:')
    print(menos_vendidos)
    print()
```

Análisis 5: costo de envío promedio por tienda

```python
envio_promedio = {}

for nombre, df in tiendas.items():
    envio_promedio[nombre] = df['Costo de envío'].mean()

df_envio = pd.DataFrame(list(envio_promedio.items()), columns=['Tienda', 'Envío Promedio'])
df_envio['Envío Promedio'] = df_envio['Envío Promedio'].round(2)
df_envio = df_envio.sort_values(by='Envío Promedio', ascending=False)
df_envio
```

crear una tabla resumen general

```python
resumen_general = pd.DataFrame({
    'Tienda': list(tiendas.keys()),
    'Ingreso Total': [df['Precio'].sum() for df in tiendas.values()],
    'Calificación Promedio': [df['Calificación'].mean() for df in tiendas.values()],
    'Envío Promedio': [df['Costo de envío'].mean() for df in tiendas.values()]
})

resumen_general['Ingreso Total'] = resumen_general['Ingreso Total'].round(2)
resumen_general['Calificación Promedio'] = resumen_general['Calificación Promedio'].round(2)
resumen_general['Envío Promedio'] = resumen_general['Envío Promedio'].round(2)

resumen_general
```

Gráficos

```python
plt.figure(figsize=(8,5))
plt.bar(resumen_general['Tienda'], resumen_general['Ingreso Total'])
plt.title('Ingreso total por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso total')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    resumen_general['Ingreso Total'],
    labels=resumen_general['Tienda'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Participación de cada tienda en el ingreso total')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(
    resumen_general['Envío Promedio'],
    resumen_general['Calificación Promedio'],
    s=100
)

for i, tienda in enumerate(resumen_general['Tienda']):
    plt.text(
        resumen_general['Envío Promedio'][i],
        resumen_general['Calificación Promedio'][i],
        tienda
    )

plt.title('Relación entre envío promedio y calificación')
plt.xlabel('Envío promedio')
plt.ylabel('Calificación promedio')
plt.show()
```
